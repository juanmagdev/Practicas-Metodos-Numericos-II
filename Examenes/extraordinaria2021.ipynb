**Ejercicio 1.**
Utilizando el método de la potencia y sus variantes, calcular toda la información posible sobre los autovalores y autovectores de la matriz:

$$A=\begin{pmatrix}
  8 & -8 & 0 & 4 & 4\\
 16 & 0 & -8 & -4 & -4\\
 18 & -14 & 2 & -2 & -2\\
 19 & -13 & -5 & 3 & -1\\
 19 & -13 & -5 & -1 & 3
\end{pmatrix}
\quad$$

Se valorará positivamente una respuesta estructurada en la forma: método utilizado, vector inicial, norma utilizada,
conclusiones.

```python
print("\nEjercicio 1")

A = array([[8, -8, 0, 4, 4], [16, 0, -8, -4, -4], [18, -14, 2, -2, -2], [19, -13, -5, 3, -1], [19, -13, -5, -1, 3]])

# print(eig(A)[0])    [-16.  16.   4.   8.   4.]

print("\nLa matriz A es:")
print(A)

X0 = array([[1], [0], [1], [0], [1]])

print("\nEl vector inicial va a ser:")
print(X0, "\n")
print(norm(X0))

#Método de la potencia
print("\n-------------------------Potencia-----------------------\n")
exito, normas, lambdas, X = potencia(A, X0, inf, 200, 1e-15)

if exito:
    print("\nConvergencia de las normas:")
    print(normas)
    print("\nConvergencia de los cocientes:")
    print(lambdas)
    print("\nconvergencia de los vectores:")
    print(X)
else:
    print(normas)

print("\nDe aqui deducimos que 4 es autovalor de A")

#Método de la potencia inversa
print("\n-------------------------Potencia Inversa-----------------------\n")
exito, normas, lambdas, X = potenciainv(A, X0, inf, 200, 1e-15)

if exito:
    print("\nConvergencia de las normas:")
    print(normas)
    print("\nConvergencia de los cocientes:")
    print(lambdas)
    print("\nconvergencia de los vectores:")
    print(X)
else:
    print(normas)

print("\nDe lo que deducimos que 1/4 es autovalor de la inv(A), luego 4 es autovalor de A.")

#Método de la potencia con desplazamiento
print("\n-------------------------Potencia Con Desplazamiento 10-----------------------\n")
exito, normas, lambdas, X = potenciades(A, X0, 10, inf, 200, 1e-15)

if exito:
    print("\nConvergencia de las normas:")
    print(normas)
    print("\nConvergencia de los cocientes:")
    print(lambdas)
    print("\nconvergencia de los vectores:")
    print(X)
else:
    print(normas)

print("\nDe aqui deducimos que -16 es autovalor de A, ya que la convergencia de los cocientes con el método")
print("de la potencia desplazado, es -6, luego restandole y sumandole el desplazamiento tenemos el -16.")

#Método de la potencia inversa con desplazamiento
print("\n-------------------------Potencia Inversa Con Desplazamiento 7-----------------------\n")
exito, normas, lambdas, X = potenciadesinv(A, X0, 7, inf, 200, 1e-15)

if exito:
    print("\nConvergencia de las normas:")
    print(normas)
    print("\nConvergencia de los cocientes:")
    print(lambdas)
    print("\nconvergencia de los vectores:")
    print(X)
else:
    print(normas)

print("\nDe aqui deducimos que 8 es autovalor de A, ya que la convergencia de los cocientes con el método de la potencia inversa")
print("con desplazamiento 7 es 1/1, luego sumandole el desplazamiento tenemos el 8.")

print("\n-------------------------Potencia Inversa Con Desplazamiento 13-----------------------\n")
exito, normas, lambdas, X = potenciadesinv(A, X0, 13, inf, 200, 1e-15)

if exito:
    print("\nConvergencia de las normas:")
    print(normas)
    print("\nConvergencia de los cocientes:")
    print(lambdas)
    print("\nconvergencia de los vectores:")
    print(X)
else:
    print(normas)

print("\nDe aqui deducimos que 16 es autovalor de A, ya que la convergencia de los cocientes con el método de la potencia inversa")
print("con desplazamiento 13 es 1/3, luego sumandole el desplazamiento tenemos el 16.")

print("\nLuego los autovalores de la matriz A son el 16, el -16, el 4 y el 8")

print("\nLos autovalores de la matriz A con la herramienta eig son:")
print(eig(A)[0])
```
