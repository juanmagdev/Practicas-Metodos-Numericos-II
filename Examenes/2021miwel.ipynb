#### Ejercicio 1.

a) Definir una función en python llamada ```trilpbanda``` que tome como argumentos de entrada $n$ y $p$ y devuelva como salida una matriz de tamaño $n\times n$, triangular inferior y $p$-semibanda, de manera que los elementos de la diagonal sean 2 y las subdiagonales sean -1 para la primera subdiagonal, -2 para la segunda subdiagonal, y así hasta ser $-p$ la subdiagonal $p$-ésima.

````python
#Ejercicio 1
print("\nEjercicio 1. Apartado ")
def trilpbanda(n, p):
    A = 2*eye(n)
    for i in range(p+1):
        A = A - i*eye(n, k = -i)
    return A    
````

b) Considerar ```A= tripanda(5,3)```, y $B=(2, 1 , -1 , -4 ,-4)^t$. Resolver el sistema $AX=B$ usando el método del ```descenso```.

````python
#Ejercicio 1
print("\nEjercicio 1. Apartado b)")
A = trilpbanda(5,3)
print("\nLa matriz A es:")
print(A)

B = array([[2], [1], [-1], [-4], [-4]])
print("\nEl segundo miembro del sistema AX = B es :")
print(B)

exito, X = descenso(A, B)

if exito:
    print("\nLa solución del sistema AX = B es:")
    print(X)
else:
    print("\nError:", X)
````

b) Se considera el sistema perturbado $\bar{A}\bar{X}=B$, donde

$$\bar{X}=X+\delta X, \qquad \bar{A}=A+\Delta A, \quad \text{siendo} \quad \Delta{A}=\verb"random.rand(4,4)/100"$$

Verificar la desigualdad 

$$\frac{\|\Delta X\|_2}{\|X+\Delta X\|_2}\leq \text{cond}_2(A) \frac{\|\Delta A\|_2}{\|A\|_2}$$
calculando todas las cantidades que aparecen en la misma.

````python
#Ejercicio 2

print("\nEjercicio 2. Apartado b)")

print("\nLa matriz dA (matriz incremento de A) es:")
random.seed(0) ## Al usar esto conseguimos que la matriz dA no cambie cada vez que corramos el programa
dA = random.rand(4,4)/100
print(dA)

## Calculemos la matriz Aper (matriz perturbada), sabiendo que Aper = A + dA
print("\nLa matriz A perturbada es:")
Aper = A - dA
print(Aper)

## Calculemos la inversa de la matriz A para calcular el condicionador_2 de A, llamemosla iA
## Para calcularla resolveremos el sistema AX=I, donde I es la identidad, mediante el método de gauss con pivot parcial

I = eye(shape(A)[0])

e, iA = gauss_pp(A, I)
print("\nLa inversa de A es:")
print(iA)

## Calculemos el condicionador_2 de A, llamemoslo cond

cond = norma_mat(A, 2)*norma_mat(iA, 2)
print("\nEl condicionador de la matriz A es cond =", cond)

## Calculemos Xper, que sabemos que es la solucion del sistema AperXper=B

e, Xper = gauss_pp(Aper, B)
print("\nLa solución del sistema perturbado es:")
print(Xper)

## Calculemos dX sabiendo que Xper = X + dX

dX = Xper - X
print("\nEl incremento de las soluciones es:")
print(dX)

## Ya tenemos caracterizados todos los elementos que estan involucrados en la desigualdad, veamos si se cumple.
## Llamemos izq al cociente de la izquierda de la desigualdad, y a los de la derecha llamemoslos der.

izq = norma_vec(dX, 2)/norma_vec(Xper, 2)

print("\nEl valor de la izquierda de la desigualdad es:", izq)

der = cond*(norma_mat(dA, 2)/norma_mat(A, 2))
print("\nEl valor de la derecha de la desigualdad es:", der)

if izq <= der :
    print("\nLa desigualdad se cumple")
else:
    print("\nLa desigualdad no se cumple")
````
